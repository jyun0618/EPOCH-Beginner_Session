### (2) 2017년과 2022년 데이터를 비교하여, 서울특별시(시도명: "서울특별시") 내에서 위치 변경이 있었던 업소들을 찾으세요. 위치 변경 여부는 경도와 위도의 변화로 판단하고, 상호명과 변화된 위치(경도, 위도) 차이를 '경도차이', '위도차이' 열을 새로 생성하여 출력하세요. (경도차이, 위도차이는 절댓값으로 출력되지 않아도 됩니다.)
* Hint: 비교 연산자 사용 (|, =! 등)

```python
seoul17=data2017[data2017['시도명']=='서울특별시']
seoul22=data2022[data2022['시도명']=='서울특별시']
```

```python
merged = seoul17.merge(seoul22,left_on='상가업소번호',right_on='상가업소번호',how='inner')
merged.head()
```

```python
merged['위도차이'] = merged['위도_x'] - merged['위도_y']
merged['경도차이'] = merged['경도_x'] - merged['경도_y']

diff = merged[(merged['위도차이'] != 0) | (merged['경도차이'] != 0)]

diff = diff[['상가업소번호','상호명_x','상호명_y','위도차이','경도차이']]
diff = diff.rename(columns={
    '상호명_x':'상호명'
})
diff

#상호명 매치 안되는 건.. ?
```

### (2) 아래와 같이 연도('year')별 평균 예산('budget')과 수익('gross')을 선 그래프로 나타내세요.
(사용한 색상은 b, forestgreen, orange, skyblue 입니다.)

1-2번.png

```python
year_group = movie.groupby('year')[['budget', 'gross']].mean().reset_index()
```

```python
plt.plot(year_group['year'], year_group['budget'], marker='.', linestyle='--', color='orange',label='Budget')
plt.plot(year_group['year'], year_group['gross'], marker='*', linestyle='-', color='forestgreen',label='Gross')
plt.hlines(1e8,1980,2020,color='blue',linestyle=':',linewidth=3)

plt.xlabel('Year')
plt.ylabel('Average Budget & Gross')
plt.title('Average Movie Budget & Gross by Year')
plt.legend(ncol=2)
plt.grid(True,axis='y',color='skyblue')
plt.show()
```
