### (3) ÏïÑÎûòÏôÄ Í∞ôÏùÄ ÏàòÎ©¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÎÑ§ Í∞ÄÏßÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨ÏÑ∏Ïöî.
* colors = ['gold', 'gray','darkgray', 'silver', 'lightgray', 'whitesmoke'] & 'coral' & 'purple' & 'green'

image.png

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sleep_counts = df["Quality of Sleep"].value_counts()
plt.pie(
    sleep_counts,
    labels=sleep_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    counterclock=False,
    colors=['gold', 'gray', 'darkgray', 'silver', 'lightgray', 'whitesmoke']
)
plt.title("Distribution of Quality of Sleep")

plt.subplot(2, 2, 2)
y = df.groupby('Age')['Sleep Duration'].mean()
plt.bar(y.index, y.values, color='coral')
plt.grid(True, axis='y', color='lightblue')
plt.xlabel('Age')
plt.ylabel('Sleep Duration (Hours)')
plt.title('Average Sleep Duration by Age')

plt.subplot(2, 2, 3)
a = df.groupby('Quality of Sleep').count()
plt.bar(a.index, a['Person ID'], color='purple')
plt.xlabel('Quality of Sleep')
plt.ylabel('Count')
plt.title('Counts of Quality of Sleep Rating')

plt.subplot(2, 2, 4)
b = df.groupby('Age')['Quality of Sleep'].mean()
plt.plot(b.index, b.values, marker='o', color='green')
plt.xlabel('Age')
plt.ylabel('Quality of Sleep')
plt.title('Quality of Sleep Trend by Age')

plt.tight_layout()
plt.show()
```
