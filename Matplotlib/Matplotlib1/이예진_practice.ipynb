```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #세번째 인자에 선의 색상과 형태를 지정하는 포맷 문자열 입력 가능. 'ro'는 빨간색(red)의 원형(o)마커를 의미
plt.axis([0,6,0,20]) #xmin,xmax,ymin,ymax
plt.show()
```

**여러 개의 그래프 그리기**

0., 5. 이렇게 .을 붙이는 이유

```python
import numpy as np

#200ms 간격으로 균일하게 샘플된 시간
t=np.arange(0.,5.,0.2)

#빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^') #s는 square, ^는 세모
plt.show()

#시작값 0, 끝값 5, 간격 0.2
```

```python
import matplotlib.pyplot as plt

# 데이터 플롯
plt.plot([1,2,3,4],[2,3,5,10],label='Price($)')
plt.plot([1,2,3,4],[3,5,9,7],label='Demand(#)')

# 축 레이블
plt.xlabel('X-Axis', labelpad=20)
plt.ylabel('Y-Axis')

# 범례 설정
plt.legend(
    loc='best',
    fontsize=8,
    frameon=True,
    shadow=True,
    facecolor='lightyellow',   # 범례 상자 배경색
    edgecolor='brown',         # 범례 상자 테두리색
    borderpad=3.0,              # 범례 테두리와 내용 간 여백
    labelspacing=2.0            # 범례 항목 간 간격
)

plt.show()
```

## **06**

**선 종류 지정**

```python
plt.plot([1,2,3],[4,4,4],'-',color='C0',label='Solid')
plt.plot([1,2,3],[3,3,3],'--',color='C0',label='Dashed')
plt.plot([1,2,3],[2,2,2],':',color='C0',label='Dotted')
plt.plot([1,2,3],[1,1,1],'-.',color='C0',label='Dash-dot')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8,3.2,0.5,5.0])
plt.legend(loc='upper center',ncol=4)
plt.show()
```

**linestyle 지정**

```python
plt.plot([1,2,3],[4,4,4],linestyle='solid',color='C0',label='Solid')
plt.plot([1,2,3],[3,3,3],linestyle='dashed',color='C0',label='Dashed')
plt.plot([1,2,3],[2,2,2],linestyle='dotted',color='C0',label='Dotted')
plt.plot([1,2,3],[1,1,1],linestyle='dashdot',color='C0',label='Dash-dot')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8,3.2,0.5,5.0])
plt.legend(loc='upper center',ncol=4)
plt.tight_layout() #그래프 요소들이 겹치지 않도록 자동으로 여백을 조정
plt.show()
```

**튜플 사용하기**

```python
x=np.arange(3)
years=['2018','2019','2020']
values=[100,400,900]

plt.bar(x,values)
plt.xticks(x,years) #눈금 레이블

plt.show()
```

```python
x=np.arange(3)
years=['2018','2019','2020']
values=[100,400,900]

plt.bar(x,values,color='y')
plt.xticks(x,years) #눈금 레이블

plt.show()
```

## **17**

**수평 막대 그래프 그리기**

```python
#bar'h'로 바꾸고 나머진 x-> y
y=np.arange(3)
years=['2018','2019','2020']
values=[100,400,900]

plt.barh(y,values,color='y')
plt.yticks(y,years) #눈금 레이블

plt.show()
```

**막대 색상 지정**

```python
y=np.arange(3)
years=['2018','2019','2020']
values=[100,400,900]
colors=['y','lightblue','C4']

plt.barh(y,values,color=colors)
plt.yticks(y,years) #눈금 레이블

plt.show()
```

**막대 높이 지정**

```python
y=np.arange(3)
years=['2018','2019','2020']
values=[100,400,900]
colors=['y','lightblue','C4']

plt.barh(y,values,color=colors,height=0.3) #높이라기보다는 두께 #근데 매우 큰 숫자를 넣으면 왜 맨 위 막대만 두꺼워질까?
plt.yticks(y,years) #눈금 레이블

plt.show()
```

**스타일 꾸미기**

```python
#bar'h'로 바꾸고 나머진 x-> y
y=np.arange(3)
years=['2018','2019','2020']
values=[100,400,900]
colors=['y','lightblue','C4']

plt.barh(y,values,color=colors,align='edge',edgecolor='#eee',linewidth=5,tick_label=years) #edge-눈금이 막대의 아래쪽 끝에.
plt.show()
```
