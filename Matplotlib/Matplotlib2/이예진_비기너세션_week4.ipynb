# **19 3차원 산점도 그리기**

```python
from mpl_toolkits.mplot3d import Axes3D #3차 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

n=100
xmin,xmax,ymin,ymax,zmin,zmax=0,20,0,20,0,50
cmin,cmax=0,2

xs=np.array([(xmax-xmin)*np.random.random_sample()+xmin for i in range(n)])
ys=np.array([(ymax-ymin)*np.random.random_sample()+ymin for i in range(n)])
zs=np.array([(zmax-zmin)*np.random.random_sample()+zmin for i in range(n)])
color=np.array([(cmax-cmin)*np.random.random_sample()+cmin for i in range(n)]) #0에서 2 사이의 값을 갖는 실수이고, 이 값을 통해 각각 다른 색 표현

fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(111,projection='3d') #111 뭐지?
ax.scatter(xs,ys,zs,c=color,marker='o',s=15,cmap='Greens')

plt.show()
```

# **24 여러 개의 그래프 그리기**

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(2,1,1) #nrow,ncol,index
plt.plot(x1,y1,'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2,1,2) #index를 각각 1,2로 입력해서 위 아래에 위치하도록 함
plt.plot(x2,y2,'.-')
plt.title('2nd Graph')
plt.xlabel('times(s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(1,2,1)
plt.plot(x1,y1,'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(1,2,2)
plt.plot(x2,y2,'.-')
plt.title('2nd Graph')
plt.xlabel('times(s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

**축 공유하기**

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

ax1=plt.subplot(2,1,1)
ex=1
plt.plot(x1,y1,'o-')
plt.ylabel('Damped oscillation')
plt.xticks(visible=False)

ax2=plt.subplot(2,1,2,sharex=ax1) #sharex,sharey 파라미터를 사용하면 두 그래프의 축 공유 가능 - 범위와 눈금이 동일하게 표시됨
ex=2
plt.plot(x2,y2,'.-')
plt.title('2nd Graph')
plt.xlabel('times(s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

# **25 컬러맵 설정하기**

# **26 텍스트 삽입하기**

**패턴의 밀도 지정하기**

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = "/" * i

    ax = plt.subplot(2, 3, i + 1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y,facecolor='deeppink', edgecolor='black')

    for bar in bars:
        bar.set_hatch(hatch_str)

plt.tight_layout()
plt.show()
```

**패턴의 두께 지정하기**

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['hatch.linewidth'] = 3 #디폴트 1

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = "+" * i

    ax = plt.subplot(2, 3, i + 1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y,facecolor='green', edgecolor='black')

    for bar in bars:
        bar.set_hatch(hatch_str)

plt.tight_layout()
plt.show()
```
